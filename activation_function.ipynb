## シグモイド関数

$$ y = \frac{1}{1+exp(-x)} $$

```
import torch
from torch import nn
import matplotlib.pylab as plt

m = nn.Sigmoid()

x = torch.linspace(-5, 5, 50)
y = m(x)

plt.plot(x, y)
plt.show()
```

## tanh (hyperbolic tangent)

$$ y = \frac{exp(x)-exp(-x)}{exp(x)+exp(-x)} $$

```
import torch
from torch import nn
import matplotlib.pylab as plt

m = nn.Tanh()

x = torch.linspace(-5,5,50)
y = m(x)

plt.plot(x,y)
plt.show()
```
